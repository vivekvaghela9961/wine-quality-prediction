# Wine Quality — Exploratory Data Analysis (EDA)

This notebook performs exploratory data analysis on the Red and White Wine Quality datasets.

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_theme(style="whitegrid")
```

## 1. Load the Datasets
We load both the red and white wine datasets, and add a `type` indicator column ('red' or 'white').

```python
# Define paths (relative to notebooks directory)
red_path = "../data/raw/winequality-red.csv"
white_path = "../data/raw/winequality-white.csv"

# Load datasets (note they are semicolon-separated)
red_df = pd.read_csv(red_path, sep=";")
white_df = pd.read_csv(white_path, sep=";")

# Add type column
red_df["type"] = "red"
white_df["type"] = "white"

# Combine datasets
df = pd.concat([red_df, white_df], ignore_index=True)
print(f"Red wine shape: {red_df.shape}")
print(f"White wine shape: {white_df.shape}")
print(f"Combined shape: {df.shape}")
df.head()
```

## 2. General Statistics & Structure

```python
df.info()
```

```python
df.describe().T
```

## 3. Missing Values & Duplicate Records

```python
print("Missing values per column:")
print(df.isnull().sum())

duplicates = df.duplicated().sum()
print(f"\nDuplicate rows: {duplicates} ({duplicates / len(df) * 100:.2f}%)")
```

## 4. Target Distribution
Let's visualize the distribution of our target variable `quality`.

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="quality", hue="type", palette={"red": "#800020", "white": "#F4E0A5"})
plt.title("Distribution of Wine Quality by Wine Type")
plt.xlabel("Quality")
plt.ylabel("Count")
plt.show()
```

## 5. Correlations
Let's look at the correlation between the chemical properties.

```python
plt.figure(figsize=(12, 10))
# Exclude the string 'type' column for correlation analysis
corr_matrix = df.drop(columns=["type"]).corr()
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", square=True, linewidths=0.5)
plt.title("Correlation Heatmap of Wine Features")
plt.show()
```

## 6. Feature Distributions and Outliers
Let's look at outlier distributions across numeric columns.

```python
features = df.columns.drop(["quality", "type"])
fig, axes = plt.subplots(nrows=4, ncols=3, figsize=(15, 15))
axes = axes.flatten()

for i, col in enumerate(features):
    sns.boxplot(data=df, x="type", y=col, ax=axes[i], palette={"red": "#800020", "white": "#F4E0A5"})
    axes[i].set_title(f"{col} distribution")
    axes[i].set_xlabel("")

# Hide any unused subplots
for j in range(len(features), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
